## Исследование надежности заемщиков

Изучение влияния семейного положения и количества детей клиента на факт возврата кредита в срок..

### Описание проекта

Кредитоспособность клиента коммерческого банка — способность заемщика полностью и в срок рассчитаться по своим долговым обязательствам (основному долгу и процентам). В отличие от его платежеспособности она не фиксирует неплатежи за истекший период или на какую-то дату, а прогнозирует способность к погашению долга на ближайшую перспективу. Уровень кредитоспособности клиента определяет степень риска банка, связанного с выдачей ссуды конкретному заемщику.

Заказчик — кредитный отдел банка. Необходимо разобраться, влияет ли семейное положение и количество детей клиента на факт погашения кредита в срок. Входные данные от банка — статистика о платёжеспособности клиентов. Результаты исследования будут учтены при построении модели кредитного скоринга — специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку.

### Описание данных

Данные сохранены в датасет data и имеют следующую структуру:
<br>
- children — количество детей в семье;
- days_employed — общий трудовой стаж в днях;
- dob_years — возраст клиента в годах;
- education — уровень образования клиента;
- education_id — идентификатор уровня образования;

- total_income — ежемесячный доход;
- purpose — цель получения кредита.

Оценку необходимо провести по следующим категориям: по доходности, по целям приобретения кредита, по статусу семейного положения и по количеству детей в семье. Зависимости данных возвращаемости кредитов по категориям визуализировать на графиках.

Проверка пропусков в данных датасета

```python
# Функция вычисления пропущенных значений в колонках датасета
def missing_values_table(df):
        # Общее количество пропусков
        mis_val = df.isnull().sum()
        
        # Доля пропущенных значений
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Создание таблицы с результатами
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Переименование колонок
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Пропуски', 1 : 'Доля, %'})
        
        # Сортировка значений по доле пропусков
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        'Доля, %', ascending=False).round(1)
        
        # Печать общей информации
        print ("Выбранный датафрейм имеет " + str(df.shape[1]) + " столбцов.\n"      
            "Из них " + str(mis_val_table_ren_columns.shape[0]) +
              " столбца имеют пропуски значений.")
        
        # Return
        return mis_val_table_ren_columns
```

```python
missing_values_table(data)
```

Пропущенные значения в датасете негативно влияют не только на исследовательский анализ данных, которые в процессе исследования могут исказить результаты, но и на перспективную разработку модели машинного обучения, для которой критичиски важна нормализованная составляющая всех данных датасета для алгоритмического поиска взаимозсвязей между признаками и целевым признаком. 

Столбец с пропущенными значениями — total_income — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца income_type. Например, у человека с типом занятости сотрудник пропуск в столбце total_income должен быть заполнен медианным доходом среди всех записей с тем же типом. Также поступим и со столбцом days_employed. Заполним пропуски в столбце days_employed медианными значениями по каждого типа занятости income_type.

Заполнение пропусков медианным значением — лучшее решение для количественных переменных. Медиана  —  это такое число в выборке, когда ровно половина элементов больше него, а другая половина —  меньше. В объеме данных в столбце могут содержаться цифры различного характера. Но когда эти цифры характеризуют несколько категорий показателей, к примеру зарплату сотрудников от инженера до начальника управления, то применять среднее арифметическое к набору данных будет неуместно и неверно. Среднее значение некорректно характеризует данные, когда некоторые значения сильно выделяются среди большинства. 

```python
for t in data['income_type'].unique():
    data.loc[(data['income_type'] == t) & (data['total_income'].isna()), 'total_income'] = \
    data.loc[(data['income_type'] == t), 'total_income'].median()
    data.loc[(data['income_type'] == t) & (data['days_employed'].isna()), 'days_employed'] = \
    data.loc[(data['income_type'] == t), 'days_employed'].median()
```

```python
# Повторная проверка пропусков значений
missing_values_table(data)
```

Обработанные пропуски в данных могли быть получены по нескольким причинам. К примеру, из-за некорректности создания веб-программистами полей ввода исходной информации. Могут данные "побиться" также при сохранении или миграции БД. Ну и элементарный человеческий фактов (забыл, плохо напечатал, схалтурил).

Заменим вещественный тип данных в столбце total_income на целочисленный с помощью метода astype()

```python
data['total_income'] = data['total_income'].astype(int)
```

### Предобработка данных days_employed

Исследование данных в столбце days_employed — общий трудовой стаж в днях.<br>
Допустим, что человек стал работать в 16 лет и завершил свою трудовую деятельность в 70 лет. Тогда идеальный общий непрерывный трудовой стаж человека составит 54 года или 19710 дней. Примем этот факт во внимание.

Построим диаграмму распределения данных в столбце days_employed

```python
fig, ax = plt.subplots(figsize=(10, 3))
sns.boxplot(x='days_employed', data=data, color='red').set(title='Распределение данных days_employed')
plt.show()
```

Эти артефакты (аномалии) — значения, которые не отражают действительность и появились по какой-то ошибке во время внесения оператором данных в БД о клиентах банка, могут сильно исказить исследовательский анализ. Прежде, чем приступить к обработке данных, проанализируем текущее состояние данных в столбцах.

```python
print('Количество строк с отрицательными значениями стажа: ', \
             (data[(data['days_employed'] < 0)]).shape[0], 'строки')
print('Количество строк с аномально высоким стажем > 54 лет: ', \
             (data[(data['days_employed'] > 19710)]).shape[0], 'строки')
print('Всего: ', (data[(data['days_employed'] < 0) + (data['days_employed'] > 19710)]).shape[0], 'строк')
print(
        'Доля аномальных значений в датасете:', \
        f"{((data[(data['days_employed'] < 0) + (data['days_employed'] > 19710)]).shape[0]/len(data)):.1%}"
)
```

Как видно, алгоритм интерфейса ввода/вывода данных о заемщиках на каком-то этапе ошибочно записал либо абсолютно все отрицательные значения стажа, либо наоборот, аномально высокие. Первым делом необходимо обработать отрицательные значения в этом столбце: заменим значения положительными с помощью метода abs() и повторно проверим результаты.

```python
data['days_employed'] = data['days_employed'].abs()
print('Количество строк с отрицательными значениями стажа: ', \
             (data[(data['days_employed'] < 0)]).shape[0], 'строк')
print('Количество строк с аномально высоким стажем > 54 лет: ', \
             (data[(data['days_employed'] > 19710)]).shape[0], 'строки')
print('Всего: ', (data[(data['days_employed'] < 0) + (data['days_employed'] > 19710)]).shape[0], 'строк')
print(
        'Доля аномальных значений после обработки:', \
        f"{((data[(data['days_employed'] < 0) + (data['days_employed'] > 19710)]).shape[0]/len(data)):.1%}"
)
print('***')
print('Минимальный трудовой стаж:', round(data['days_employed'].min(), 1), 'дней')
print('Максимальный трудовой стаж:', round(data['days_employed'].max(), 1), 'дней (', round(data['days_employed'].max()/365,), 'лет )')
```

Для каждого типа занятости выведем медианное значение трудового стажа days_employed в днях

```python
# Вывод медианного значения трудового стажа
data.groupby('income_type')['days_employed'].agg('median')
```

У двух типов (безработные и пенсионеры) получились аномально высокие значения. Исправить такие значения можно методом замены на адекватные значения относительно каждой категории клиентов.

```python
print('Количество пенсинеров с аномальным стажем: ', (data[(data['days_employed'] > 19710) & 
                    (data['income_type'] == 'пенсионер')]).shape[0], 'строк')
print('Количество безработных с аномальным стажем: ', (data[(data['days_employed'] > 19710) & 
                    (data['income_type'] == 'безработный')]).shape[0], 'строк')
```

Будем считать, что данная категория заемщиков изначально имеет самый низкий уровень кредитоспособности и не несет в себе особой ценности для банка. Поэтому заменим аномалии по стажу работы на максимальные значения датасета для пенсионеров и на минимальные - для безработных (исключив из условия выборки сами эти категории).

```python
# Перезапись данных в столбце days_employed по условиям
data.loc[(data['days_employed'] > 19710) & (data['income_type'] == 'пенсионер'), 'days_employed'] = \
     data['days_employed'].where((data['income_type'] != 'пенсионер') & (data['income_type'] != 'безработный')).max()
data.loc[(data['days_employed'] > 19710) & (data['income_type'] == 'безработный'), 'days_employed'] = \
     data['days_employed'].where((data['income_type'] != 'пенсионер') & (data['income_type'] != 'безработный')).min()
```

```python
data['days_employed'] = data['days_employed'].abs()
print('Количество строк с отрицательными значениями стажа: ', \
             (data[(data['days_employed'] < 0)]).shape[0], 'строк')
print('Количество строк с аномально высоким стажем > 54 лет: ', \
             (data[(data['days_employed'] > 19710)]).shape[0], 'строки')
print('Всего: ', (data[(data['days_employed'] < 0) + (data['days_employed'] > 19710)]).shape[0], 'строк')
print(
        'Доля аномальных значений после обработки:', \
        f"{((data[(data['days_employed'] < 0) + (data['days_employed'] > 19710)]).shape[0]/len(data)):.1%}"
)
print('***')
print('Минимальный трудовой стаж:', round(data['days_employed'].min(), 1), 'дней')
print('Максимальный трудовой стаж:', round(data['days_employed'].max(), 1), 'дней (', round(data['days_employed'].max()/365,), 'лет )')
```

Очень много однотипных значений. Создадим функцию, которая на основании данных из столбца purpose сформирует новый столбец purpose_category, в который войдут следующие категории:

 - операции с автомобилем,
 - операции с недвижимостью,
 - проведение свадьбы,
 - получение образования.
 
Например, если в столбце purpose находится подстрока 'на покупку автомобиля', то в столбце purpose_category должна появиться строка 'операции с автомобилем'.

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

```python
# Выведем перечень уникальных значений столбца children
print(sorted(data['purpose_category'].unique()))
```

### Зависимость возврата кредита от количества детей

*Есть ли зависимость между количеством детей и возвратом кредита в срок?*

Для того, чтобы выяснить зависимость между кол-вом детей и возвратом кредита в срок, воспользуемся встроенной в библиотеку pandas функцией groupby. Сгруппируем строки по столбцу 'children' и найдем сумму значений по каждой группе по столбцу 'debt'. Это позволит сразу визуально определить соотношение объемов задолженности по каждой категории  кредита столбцу children. Результат сохраним в таблицу data_credit

```python
#data_credit = data.groupby('children')['debt'].sum()
data_credit = data.groupby(['children'])['debt'].sum().reset_index()

"""
Чтобы определить в цифрах зависимость задолженности 
от кол-во детей в семье создадим вторую таблицу data_child 
для вывода общего кол-ва детей по каждой категории многодетности.
"""

data_child = data.groupby(['children'])['children'].count()
data_child = pd.DataFrame({'children':data_child.index, 'count':data_child.values})

"""
Теперь, чтобы рассчитать долю задолженностей по каждой категории 
многодетности склеим две идентичные таблицы data_credit и data_child 
в одну таблицу result
"""
#result = pd.concat([data_child, data_credit])
result = data_credit.merge(data_child, how='inner', on='children')

# и добавим столбец 'debt level' с форматом отображения в %
result['debt_level_percent'] = (result['debt'] / result['count']).apply('{:.2%}'.format)

# и добавим столбец 'debt level' с форматом отображения в %
result['debt_level'] = (result['debt'] / result['count'])

# и добавим столбец 'debt level all'
result['debt_level_all'] = result['debt'] / result['debt'].sum()

result_tab = result.copy()
result_tab = result_tab[['children', 'count', 'debt', 'debt_level_percent']]

# Изменяем названия столбцов на понятные
result_tab.columns = ['Число детей', 'Всего должников', 'Объем долгов', '%, Доля']
```

Чтобы провести анализ зависимости между уровнем дохода и возвратом кредита в срок, удобнее всего будет произвести категоризацию количественных данных. Соотношение доходов среди заемщиков 1 / 110. Слишком большой разброс доходов. Необходимо произвести категоризацию данных так, чтобы она максимально качественно отразилась на результатах исследования. Когда мы делим на категории количественные данные, то лучше их делать примерно одинаковыми. Для этого нам поможет метод qcut().

В документации qcut описывается как "функция дискретизации на основе квантилей". По сути, это означает, что qcut пытается разделить базовые данные на интервалы равного размера. Функция определяет интервалы с использованием процентилей на основе распределения данных, а не фактических числовых границ интервалов.

Типичным вариантом использования является сохранение результатов разбиения в исходном фрейме данных (dataframe) для будущего анализа. Одна из проблем, связанных с этим подходом, заключается в том, что имена интервалов сложно объяснить конечному пользователю. Поэтому нам необходимо явно назвать интервалы, чтобы их было легче интерпретировать.

```python
# Определим интервалы
pd.qcut(total_income_debt['total_income'], q=5, precision=0).head(1)
```

```python
"""
По идее из полученных разбитых на 5 частей категорий доходности  
можно было бы разбить доходность заемщиков вручную функцией if-else.

def income_category(income):
    if income <= 100000:
        return 'До 98527 руб.'
    if income > 100000 and income <= 200000:
        return 'От 98528 до 132137 руб.'
    if income > 200000 and income <= 300000:
        return 'От 132138 до 161371 руб.'
    if income > 300000 and income <= 400000:
        return 'От 161372 до 214550 руб.'
    if income > 400000 and income <= 500000:
        return 'Свыше 214550 руб.'
    return
    
С помощью функции и метода apply() надо было потом добавить в анализируемую таблицу 
total_income_debt новый столбец с категориям доходности. Для этого снова 
потребовалось бы создание столбца

total_income_debt['total_income_category'] = ( total_income_debt['total_income']
                                              .apply(income_category)
                                             )
total_income_debt
    
Но мы воспользовались инструментом qcut и решили 
нашу задачу гораздо эффективнее и быстрее!
"""

bin_labels_5 = ['До 98527 руб.', 'От 98528 до 132137 руб.', 
                'От 132138 до 161371 руб.', 'От 161372 до 214550 руб.', 
                'Свыше 214550 руб.']

total_income_debt['total_income_category'] = pd.qcut(total_income_debt['total_income'],
                              q=[0, .2, .4, .6, .8, 1],
                              labels=bin_labels_5)
```

### Влияние целей кредита на его возврат

Как разные цели кредита влияют на его возврат в срок?

Ранее мы уже сформировали новый столбец purpose_category, в котором скатегорировали все цели на 4 основные категории:
- операции с недвижимостью;
- операции с автомобилем;
- получение образования;
- проведение свадьбы.

Анализ влияния целей кредита на его возврат в срок проведем относительно данной категории целей. По столбцу 'purpose_category'.

```python
data['purpose_category'].unique()
```

Зависимость между целями кредита и их возвратом в срок, будем строить по столбцам 'purpose_category' и 'debt'. Снова воспользуемся встроенной в библиотеку pandas функцией groupby. Сгруппируем строки по столбцу 'purpose_category' и найдем для каждой группы сумму значений по столбцу 'debt'. Результат сохраним в таблицу purpose_debt.

```python
purpose_debt = data.groupby('purpose_category')['debt'].sum()
# Проверяем, что получилось в таблице purpose_debt
purpose_debt.head()
```

```python
# Cоздадим вторую таблицу purpose_count для вывода общего кол-ва строк по каждой категории цели.
purpose_count = data.groupby('purpose_category')['purpose_category'].count()
purpose_count.head()
```

Для того, чтобы проанализировать зависимость целями кредита и их возвратом в срок рассчитаем долю возврата кредита в срок по каждой категории цели. Для этого склеим две идентичные таблицы purpose_debt и purpose_count в одну таблицу result.

```python
purpose_debt_sum = pd.DataFrame({'purpose_category':purpose_debt.index, 'debt':purpose_debt.values})
purpose_debt_count = pd.DataFrame({'purpose_category':purpose_count.index, 'count':purpose_count.values})

result = purpose_debt_sum.merge(purpose_debt_count, how='inner', on='purpose_category')

# и добавим столбец 'debt level' с форматом отображения в %
result['debt_level_percent'] = (result['debt'] / result['count']).apply('{:.2%}'.format)

# и добавим столбец 'debt level' с форматом отображения в %
result['debt_level'] = (result['debt'] / result['count'])

# и добавим столбец 'debt level all'
result['debt_level_all'] = result['debt'] / result['debt'].sum()

result_tab = result.copy()
result_tab = result_tab[['purpose_category', 'count', 'debt', 'debt_level_percent']]

# Изменяем названия столбцов на понятные
result_tab.columns = ['Категория доходов', 'Всего должников', 'Объем долгов', '%, Доля']
```

```python
# make figure and assign axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)

# pie chart parameters
a = round((result['count'].sum() - result['debt'].sum())/result['count'].sum(),3)
b = round(result['debt'].sum()/result['count'].sum(),3)
#overall_ratios = round((result_tab['Объем долгов']/result_tab['Всего заемщиков'].sum()),3)
#labels = result_tab['Категория доходов'].tolist()
overall_ratios = [b, a]
labels = ['С задолженностью', 'Без долгов']
explode = [0.1, 0]
# rotate so that first wedge is split by the x-axis
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.1f%%', startangle=angle,
                    labels=labels, explode=explode)

ax1.set_title('Влияние целей кредита\n' 
              'и его возврат в срок')
# bar chart parameters
age_ratios = result['debt_level_all']
age_labels = result['purpose_category'].astype(str)
bottom = 1
width = .2

# Adding from the top matches the legend.
for j, (height, label) in enumerate(reversed([*zip(age_ratios, age_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label,
                 alpha=0.1 + 0.18 * j
                )
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax2.set_title('Доля задолженностей')
ax2.legend(bbox_to_anchor=( 1.4 , 1), loc='upper right')
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

# use ConnectionPatch to draw lines between the two plots
theta1, theta2 = wedges[0].theta1, wedges[0].theta2
center, r = wedges[0].center, wedges[0].r
bar_height = sum(age_ratios)

# draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(-width / 2, bar_height), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(2)
ax2.add_artist(con)

# draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(-width / 2, 0), coordsA=ax2.transData,
                      xyB=(x, y), coordsB=ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(2)

plt.yscale("symlog")
plt.show()

print('Таблица возвращаемости кредитов')
result_tab.sort_values(by='%, Доля', ascending=False).reset_index(drop=True).style.hide_index()
```

Анализ зависимости целей кредита на их возвращаемость наглядно показывает, что самые благонадежные среди заемщиков - это те, кто приобретает недвижимость и скрепляет законные узы брака. Исследуя зависимости возврата кредитов в срок по разным условиям, пришел к выводу, что самые надежные заемщики - это семейные люди, либо планирующие вступить в законный брак.
